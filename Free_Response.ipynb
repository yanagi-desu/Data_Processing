```python
#http://www.nltk.org/book-jp/ch12.html#id32
import scipy
import nltk
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from nltk.probability import FreqDist
import nagisa
from matplotlib.pyplot import figure
import matplotlib.font_manager as fm
jprop = fm.FontProperties(fname='YasashisaGothicBold-V2.otf')
```

```python
survey_raw = pd.read_csv("Raw_survey_data.csv")
market_worries = survey_raw["q6t"]
market_likes = survey_raw["q7t"]
market_worries[2]
meaningless = ["なし","特になし","とくになし","ありません","ない","特に無い"]
removed_nones_timeline_worries = [each for each in market_worries if each not in meaningless]
removed_nones_timeline_likes = [each for each in market_likes if each not in meaningless]
print(len(removed_nones_timeline_worries))
```

```python
# words = nagisa.filter("".join(removed_nones_timeline_worries), filter_postags=['名詞', '形状詞'])
words = nagisa.extract("".join(removed_nones_timeline_worries), extract_postags=['名詞', '形状詞'])
```

```python
#Market Worries plt
fdist = FreqDist(words.words)
top = fdist.most_common(33)
print(top)
remove = ["不安","よう","気"]
for each in top:
    if each[0] in remove:
        top.remove(each)
# graph = fdist.plot(30,cumulative=False)
x = [i[0] for i in top]
y = [i[1] for i in top]
# Plot the graph
figure(figsize=(27, 10), dpi=80)
plt.bar(x,y)
plt.xticks(x, fontproperties=jprop, fontsize=14)
plt.title("Top 30 key words that appeared in negative opionion towards Market section", fontproperties=jprop, fontsize=16)
plt.show()
```

```python
#Market Likes plt
words_likes = nagisa.extract("".join(removed_nones_timeline_likes), extract_postags=['名詞', '形状詞'])
fdist = FreqDist(words_likes.words)
top = fdist.most_common(38)
print(top)
remove = ["不安","よう","気","そう","点","ところ","必要","可能"]
for each in top:
    if each[0] in remove:
        top.remove(each)
# graph = fdist.plot(30,cumulative=False)
x = [i[0] for i in top]
y = [i[1] for i in top]
# Plot the graph
figure(figsize=(27, 10), dpi=80)
plt.bar(x,y)
plt.xticks(x, fontproperties=jprop, fontsize=14)
plt.title("Top 30 key words that appeared in Postive opionion towards Market section", fontproperties=jprop, fontsize=16)
plt.show()
```

```python
timeline_worries = survey_raw["q9t"]
timeline_likes = survey_raw["q10t"]
timeline_worries[2]
meaningless = ["なし","特になし","とくになし","ありません","ない","特に無い"]
removed_nones_timeline_worries = [each for each in timeline_worries if each not in meaningless]
removed_nones_timeline_likes = [each for each in timeline_likes if each not in meaningless]
```

```python
#Market Worries plt
words = nagisa.extract("".join(removed_nones_timeline_worries), extract_postags=['名詞', '形状詞'])
fdist = FreqDist(words.words)
top = fdist.most_common(30)
print(top)
remove = ["不安","よう","気","メディア","こと","そう"]
for each in top:
    if each[0] in remove:
        top.remove(each)
# graph = fdist.plot(30,cumulative=False)
x = [i[0] for i in top]
y = [i[1] for i in top]
# Plot the graph
figure(figsize=(35, 10), dpi=90)
plt.bar(x,y)
plt.xticks(x, fontproperties=jprop, fontsize=12)
plt.title("Top 30 key words that appeared in negative opionions towards Timeline section", fontproperties=jprop, fontsize=16)
plt.show()
```

```python
#Timeline Worries plt
words = nagisa.extract("".join(removed_nones_timeline_likes), extract_postags=['名詞', '形状詞'])
fdist = FreqDist(words.words)
top_likes = fdist.most_common(30)

remove = ["不安","よう","気","メディア","こと","そう"]
for each in top_likes: 
    if each[0] in remove:
        top_likes.remove(each)
top_likes.pop(0)
# graph = fdist.plot(30,cumulative=False)
x = [i[0] for i in top_likes]
print(x)
y = [i[1] for i in top_likes]
# Plot the graph
figure(figsize=(30, 10), dpi=90)
plt.bar(x,y)
plt.xticks(x, fontproperties=jprop, fontsize=12)
plt.title("Top 30 key words that appeared in positive opionions towards Timeline section", fontproperties=jprop, fontsize=16)
plt.show()

```
